```python
years = range(2015, 2023)
dfs = [df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022]
```

# Pivot
## Happiness Score

```python
pivoted_data = df_all.pivot_table(index='Year', columns='Country', values='Happiness Score')
```

# Visualization
## Happiness Score

```python
plt.figure(figsize=(30, 8))

# Define the number of countries and calculate the width of each bar so they fit next to each other
countries = ['Switzerland', 'Turkey', 'Togo', 'United States']
n_countries = len(countries)
bar_width = 0.25  # Adjust this value as needed to change the width of the bars

# Create an array with the positions of each group of bars along the x-axis
x = np.arange(len(pivoted_data.index))  # the label locations

# Loop through the countries, creating a bar for each one, slightly offset
for i, country in enumerate(countries):
    bars = plt.bar(x + i * bar_width, pivoted_data[country], width=bar_width, label=country)
    
    # Add two sets of labels for each bar: one for the country and one for the value
    for bar in bars:
        # Label with the country name
        plt.text(bar.get_x() + bar.get_width() / 2, 0, country,
                 ha='center', va='bottom', rotation=90, fontsize=9)
        
        # Label with the data value above the bar
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.2f}',
                 ha='center', va='bottom', fontsize=9)

plt.title('Happiness Scores')
plt.xlabel('Year')
plt.ylabel('Happiness Score')
# Adjust the x-ticks to center them under the groups of bars
plt.xticks(x + bar_width * (n_countries - 1) / 2, pivoted_data.index)
plt.legend()

plt.show()
```
