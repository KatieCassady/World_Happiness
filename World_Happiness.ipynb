# Dictionary Creation

```python
years = range(2015, 2023)
dfs = [df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022]
```

# Combined Dataframes

```python
for year, df in zip(years, dfs):
    df['Year'] = year

df_all = pd.concat(dfs)
```

# Process Verification

```python
df_all.info()
```

# Replace N/A or NAN

```python
df_all.replace('N/A', np.nan, inplace=True)
df_all.fillna(0, inplace=True)
```

# Pivot

```python
pivoted_data = df_all.pivot_table(index='Year', columns='Country', values='Happiness Score')
```

# Visualization

```python
plt.figure(figsize=(30, 8))

# Define the number of countries and calculate the width of each bar so they fit next to each other
countries = ['Switzerland', 'Turkey', 'Togo']
n_countries = len(countries)
bar_width = 0.25  # Adjust this value as needed to change the width of the bars

# Create an array with the positions of each group of bars along the x-axis
x = np.arange(len(pivoted_data.index))  # the label locations

# Loop through the countries, creating a bar for each one, slightly offs
for i, country in enumerate(countries):
    plt.bar(x + i * bar_width, pivoted_data[country], width=bar_width, label=country)

plt.title('Pre-COVID Happiness Scores')
plt.xlabel('Year')
plt.ylabel('Happiness Score')
plt.xticks(x + bar_width / n_countries, pivoted_data.index)  # Center the x-axis ticks between the groups of bars
plt.legend()

plt.show()
```
