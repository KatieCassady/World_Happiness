```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_2015=pd.read_csv('World_Happiness_Index\\2015.csv')
df_2016=pd.read_csv('World_Happiness_Index\\2016.csv')
df_2017=pd.read_csv('World_Happiness_Index\\2017.csv')
df_2018=pd.read_csv('World_Happiness_Index\\2018.csv')
df_2019=pd.read_csv('World_Happiness_Index\\2019.csv')
df_2020=pd.read_csv('World_Happiness_Index\\2020.csv')
df_2021=pd.read_csv('World_Happiness_Index\\2021.csv')
df_2022=pd.read_csv('World_Happiness_Index\\2022.csv')
```

# Drop Operation

```python
df_2015.drop(['Standard Error', 'Region', 'Dystopia Residual'], axis = 1, inplace=True)
df_2016.drop(['Lower Confidence Interval', 'Upper Confidence Interval', 'Region', 'Dystopia Residual'], axis = 1, inplace=True)
df_2017.drop(['Whisker.high', 'Whisker.low', 'Dystopia Residual'], axis = 1, inplace=True)
df_2020.drop(['Regional indicator', 'Standard error of ladder score', 'upperwhisker', 'lowerwhisker', 'Ladder score in Dystopia', 'Explained by: Log GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices', 
'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Dystopia + residual'], axis = 1, inplace=True)
df_2021.drop(['Regional indicator', 'Standard error of ladder score', 'upperwhisker', 'lowerwhisker', 'Ladder score in Dystopia', 'Explained by: Log GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices', 
'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Dystopia + residual'], axis = 1, inplace=True)
df_2022.drop(['Whisker-high', 'Whisker-low', 'Dystopia Residual'], axis = 1, inplace=True)
```

# Dictionary Creation

```python
years = range(2015, 2023)
dfs = [df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022]
```

# Combined Dataframes

```python
for year, df in zip(years, dfs):
    df['Year'] = year

df_all = pd.concat(dfs)
```

# Process Verification

```python
df_all.info()
```

# Replace N/A or NAN

```python
df_all.replace('N/A', np.nan, inplace=True)
df_all.fillna(0, inplace=True)
```

# Process Validation

```python
df_all.info()
```

# Pivot
## Happiness Score

```python
pivoted_data = df_all.pivot_table(index='Year', columns='Country', values='Happiness Score')
```

# Visualization
## Happiness Score

```python
plt.figure(figsize=(30, 8))

# Define the number of countries and calculate the width of each bar so they fit next to each other
countries = ['Switzerland', 'Turkey', 'Togo', 'United States']
n_countries = len(countries)
bar_width = 0.25  # Adjust this value as needed to change the width of the bars

# Create an array with the positions of each group of bars along the x-axis
x = np.arange(len(pivoted_data.index))  # the label locations

# Loop through the countries, creating a bar for each one, slightly offset
for i, country in enumerate(countries):
    bars = plt.bar(x + i * bar_width, pivoted_data[country], width=bar_width, label=country)
    
    # Add two sets of labels for each bar: one for the country and one for the value
    for bar in bars:
        # Label with the country name
        plt.text(bar.get_x() + bar.get_width() / 2, 0, country,
                 ha='center', va='bottom', rotation=90, fontsize=12)
        
        # Label with the data value above the bar
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.2f}',
                 ha='center', va='bottom', fontsize=9)

plt.title('Happiness Scores')
plt.xlabel('Year')
plt.ylabel('Happiness Score')
# Adjust the x-ticks to center them under the groups of bars
plt.xticks(x + bar_width * (n_countries - 1) / 2, pivoted_data.index)
plt.legend()

plt.show()
```

# Conclusion

I had hypothesised that over the course of COVID with all the restrictions, illness, death, & trauma, that happiness, trust, and freedom scores would be lower, if not drastically lower. However, at least per this global survey, that is not the case. Across the selected countries, the happiness level stayed relatively consistent pre-, during, and post-COVID. Most surprising to me was that the trust in goverment and freedom to make life choices went up in all 4 sampled countries. And in all but Switzerland, who stayed relatively constant in trust, the trust in government and freedom to make life choices numbers went up drastically during COVID to then fall sharply post-COVID. With any survey there is always the knowlede that you are only getting the opinion of people willing to answer the survey, but the results still surprised me.
